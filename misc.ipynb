```python
import pandas as pd
```

```python
evaldf = pd.read_csv("eval.csv")
slangdf = pd.read_csv("slang_data.csv")
```

```python
eval_dict = dict(zip(evaldf['slang'], evaldf['example_3']))
eval_dict
```

```python
slang_dict = slangdf.to_dict(orient='records')
slang_dict[0]
```

```python
# For each entry in eval_dict, add a new row to slangdf with:
# - slang: the key from eval_dict
# - definition: the existing definition from slangdf (matched by slang)
# - example: the value from eval_dict

new_rows = []
for slang_word, example in eval_dict.items():
    # Find the definition for this slang word in slangdf
    definition = slangdf[slangdf['slang'] == slang_word]['definition'].values
    if len(definition) > 0:
        new_rows.append({
            'slang': slang_word,
            'definition': definition[0],
            'example': example
        })

# Add new rows to slangdf
slangdf_updated = pd.concat([slangdf, pd.DataFrame(new_rows)], ignore_index=True)
slangdf_updated = slangdf_updated.sort_values('slang', ascending=False).reset_index(drop=True)
slangdf_updated
```

```python
slangdf_updated.to_csv("dataset.csv", index=False)
```

```python
# Read dataset.csv
dataset_df = pd.read_csv("dataset.csv")

# Get slang words from dataset
dataset_slang = set(dataset_df['slang'].unique())

# Drop example_3 column from evaldf
evaldf = evaldf.drop(columns=['example_3'])

# Keep only entries where slang is in both dataset and evaldf
evaldf_filtered = evaldf[evaldf['slang'].isin(dataset_slang)]

evaldf_filtered
```

```python
evaldf_filtered.to_csv("eval.csv", index=False)
```
